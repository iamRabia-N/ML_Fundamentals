## Basics
Binary classification involves categorizing data into one of two classes or categories based on its features or attributes. It's a fundamental task in ML where the goal is to predict a binary outcome. At its core, binary classification deals with decision-making based on available information. Each data point is assigned a label typically represented as 0 or 1, true or false, positive or negative etc.



## Goal 
In a binary classification, the task is to learn a mapping from input features to binary labels. Given a dataset with labeled examples, the goal is to build a model that can accurately predict the class labels of unseen instances.



## Algorithms
Several algorithms are commonly used for binary classification including:
- Logistic Regression (ML)
- Decision Trees (ML)
- Random Forests (Ensemble ML)
- Support Vector Machines (ML)
- Gradient Boosting Machines (Ensemble ML)
- Neural Networks (DL)



## Evaluation Metrics
To assess the performance of a binary classification model, various evaluation metrics are used including:
- Accuracy
- Precision
- Recall
- F1 Score
- ROC curve
- AUC



## Applications
Binary classification finds applications across various domains such as:
- Spam detection in emails (Classifying emails as spam or not spam)
- Medical diagnosis (Determining whether a patient has a certain disease or not)
- Fraud detection (Identifying fraudulent transactions in financial systems)
- Sentiment analysis (Analyzing text data to determine whether it expresses positive or negative sentiment)
